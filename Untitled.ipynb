```python
"""2.sum and product of numbers upto using for and while loop"""

#for loop
i=0
sum=0
n=int(input())
for i in range(0,n+1):
  sum+=i
  i+=1
sum
```

```python
#while loop
i=1
sum=0
n=int(input())
while i<=n:
  sum+=i
  i+=1
sum
```

```python
"""3.electricity bill"""

nu=int(input())
if(nu>0):
  if(nu>0 and nu<=100):
    p=nu*4.5
  if(nu>100 and nu<=200):
    p=100*4.5+100*6
  if(nu>200 and nu<=300):
    p=100*4.5+100*6+100*10
  if(nu>300):
    p=(100*4.5+100*6+100*10)+(nu-300)*20
print(p)
```

```python
#using for loop
l=list(range(1,101))
l
l1=[]
for i in l:
  s=i*i*i
  if s%4==0 or s%5==0:
    l1.append(s)
l1
```
